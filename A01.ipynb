# Regresiones Polinomiales – Batería de Tableta


**Regresión polinomial de grado 1**


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
bateria = pd.read_csv(r"C:\\Users\\elisa\\lab\bateria_tableta.txt")
salarios = pd.read_csv("Salarios_minimos.csv")
```

```python
x = bateria[["carga"]]   
y = bateria["duracion"]
```

```python
modelo1 = LinearRegression()
modelo1.fit(x, y)
```

```python
r2_1 = modelo1.score(x, y)
print("R² grado 1:", r2_1)
```

```python
intercepto_1 = modelo1.intercept_
pendiente_1 = modelo1.coef_[0]

print("Ecuación grado 1:")
print(f"y = {intercepto_1:.4f} + {pendiente_1:.4f}x")
```

```python
print("R² grado 1:", r2_1)
```
