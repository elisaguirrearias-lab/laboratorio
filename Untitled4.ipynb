```python
import pandas as pd
import numpy as np

data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
datos=pd.read_excel(r"C:\Users\elisa\Downloads\Boston Housing Dataset 1978.xlsx")
datos.head(20)
```

```python
dfscaled=datos.copy()
for col in dfscaled.columns :
    minn=dfscaled[(col)].min()
    dfscaled[col]=(datos[col]-minn)
    maxx=dfscaled[(col)].max()
    dfscaled[col]/=maxx
```

```python
   dfscaled.describe()
```

```python
dfstd=datos.copy()
for col in dfstd.columns:
    media=dfstd[col].mean()
    std=dfstd[col].std()
    dfstd[col]=(dfstd[col]-media)/std
```

```python
dfstd.describe()
```

```python
import matplotlib.pyplot as plt
plt.hist(dfstd["CRIM"], bins=30)
plt.xlabel("CRIM estandarizado")
plt.ylabel("Frecuencia")
plt.title("Histograma de CRIM (escalamiento estndar)")
plt.show()
```
