```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from scipy import stats
import statsmodels.api as sm
```

```python
df= pd.read_excel(r"C:\Users\elisa\Downloads\aguasfeo.xlsx")
```

```python
df.columns = df.columns.str.strip()
df
```

**Regresión lineal**

```python
df1 = pd.DataFrame({
    "Tamaño": (df["Tamaño"] == "grande").astype(int),
    "Hielo": (df["Hielo"] == "con ").astype(int),
    "Kahlúa": (df["Sabor"] == "kahlúa").astype(int),
    "Horchata": (df["Sabor"] == "horchata").astype(int)
})

df1.head(10)
```

```python
x1 = df1[["Tamaño","Hielo","Kahlúa","Horchata"]]
x = sm.add_constant(x1)
y = df[["Score"]].values.reshape(-1,1)
```

```python
xd = sm.add_constant(x)
model = sm.OLS(y,xd)
results = model.fit()
results.summary()
```

**Parte 2**

```python
df2 = pd.DataFrame({
    "Tamaño": ((df["Tamaño"] == "chica").astype(int) * 2 - 1),
    "Hielo": ((df["Hielo"] == "con ").astype(int) * 2 - 1),
    "Kahlúa": (df["Sabor"] == "kahlúa").astype(int),
    "Horchata": (df["Sabor"] == "horchata").astype(int)
})
df2.head(10)
```

```python
X2 = sm.add_constant(df2)
y2 = df["Score"]
```

```python
modelo2 = sm.OLS(y2, X2)
resultados2 = modelo2.fit()
resultados2.summary()
```

**conclusion**

En las dos regresiones el R² es de 83.7%, lo que significa que el modelo explica el 83.7% del comportamiento del score. Cuando cambiamos la codificación de (0,1) a (-1,1), los coeficientes sí cambian, pero el ajuste del modelo no cambia. Es decir, sigue siendo el mismo modelo, solo que cambia la forma en la que se interpretan los valores.
