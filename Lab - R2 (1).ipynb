## Regresión polinomial

Observa la siguiente gráfica.

```python
import matplotlib.pyplot as plt; import numpy as np
%matplotlib inline
np.random.seed(0)
x = np.linspace(-3, 3, 100); y = 1 + 4*x**2 + np.random.normal(0, 1, 100)
plt.figure(); plt.xlabel("x"); plt.ylabel("y")
plt.scatter(x, y); plt.grid()
```

Utiliza `sklearn` para realizar una regresión lineal simple ($\hat{y} = \hat{\beta_0} + \hat{\beta_1}x $)

```python
from sklearn.linear_model import LinearRegression
```

```python
x = x.reshape(-1,1)
```

```python
modelo=LinearRegression()
modelo.fit(x,y)
```

```python
intercepto=modelo.intercept_[0]
pendiente=modelo.coef_[0][0]
print("Ecuación grado 1:")
print (f"y = {intercepto}  {pendiente}x")
```

Crea una gráfica donde se muestren las observaciones como una nube de puntos y la línea que ajustaste con regresión. Calcula el métrico $R^2$.

```python
ypred=modelo.predict(x)
```

```python
plt.figure()
plt.scatter(x, y, label="Observaciones")
plt.plot(x, ypred, label="Recta de regresión")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.show()
```

```python
r2=modelo.score(x,y)
print("R2:",r2)
```

Podemos ver que el ajuste no es correcto, pues parece que tiene un comportamiento cuadrático y estamos intentando ajustar una línea recta.

En lugar de hacer una regresión lineal simple, quizás podamos extender el modelo para que tenga varios factores dentro:

$$ Y = \beta_0 + \beta_1 X + \beta_2 X^2 + ... + \beta_p X^p + \epsilon$$

Ajusta un modelo cuadrático con `sklearn` ($\hat{y} = \hat{\beta_0} + \hat{\beta_1}x + \hat{\beta_1}x^2 $) y calcula su $R^2$.

```python
x2=x**2
```

```python
X2 = np.hstack((x,x2))
```

```python
modelo2=LinearRegression()
modelo2.fit(X2,y)
a=modelo2.intercept_[0]
b1,b2=modelo2.coef_[0]
print("Ecuación grado 2:")
print(f"y={a} {b1}x + {b2}x^2")
```

```python
r2_2=modelo2.score(X2,y)
print("R2:",r2_2)
```

Crea una gráfica con las observaciones y la nueva curva que ajustaste.

```python
y2_pred = modelo2.predict(X2)
plt.figure()
plt.scatter(x, y, label="Observaciones")
plt.plot(x, y2_pred, label="Modelo grado 2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.show()
```

Ajusta modelos de grados 3, 5, 7, 9, y 11. Calcula su $R^2$.

```python
x3 = x**3
X3 = np.hstack((x,x2,x3))
modelo3 = LinearRegression()
modelo3.fit(X3, y)
y3_pred = modelo3.predict(X3)
r2_3 = modelo3.score(X3, y)
print("R² grado 3:", r2_3)
plt.figure()
plt.scatter(x, y)
plt.plot(x, y3_pred)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
x4 = x**4
x5 = x**5
X5 = np.hstack((x, x2, x3, x4, x5))

modelo5 = LinearRegression()
modelo5.fit(X5, y)

y5_pred = modelo5.predict(X5)
r2_5 = modelo5.score(X5, y)

print("R² grado 5:", r2_5)

plt.figure()
plt.scatter(x, y)
plt.plot(x, y5_pred)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
x6 = x**6
x7 = x**7
X7 = np.hstack((x, x2, x3, x4, x5, x6, x7))

modelo7 = LinearRegression()
modelo7.fit(X7, y)

y7_pred = modelo7.predict(X7)
r2_7 = modelo7.score(X7, y)

print("R² grado 7:", r2_7)

plt.figure()
plt.scatter(x, y)
plt.plot(x, y7_pred)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
x8 = x**8
x9 = x**9
X9 = np.hstack((x, x2, x3, x4, x5, x6, x7, x8, x9))

modelo9 = LinearRegression()
modelo9.fit(X9, y)

y9_pred = modelo9.predict(X9)
r2_9 = modelo9.score(X9, y)

print("R² grado 9:", r2_9)

plt.figure()
plt.scatter(x, y)
plt.plot(x, y9_pred)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

Crea una gráfica por cada modelo con las observaciones y su curva.

```python
x10 = x**10
x11 = x**11
X11 = np.hstack((x, x2, x3, x4, x5, x6, x7, x8, x9, x10, x11))

modelo11 = LinearRegression()
modelo11.fit(X11, y)

y11_pred = modelo11.predict(X11)
r2_11 = modelo11.score(X11, y)

print("R² grado 11:", r2_11)

plt.figure()
plt.scatter(x, y)
plt.plot(x, y11_pred)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

**¿Cuál modelo consideras que es mejor? ¿Por qué?**

Debido a que el modelo se construyó a partir de una función cuadrática \(y = 1 + 4x^2 + r\), el modelo que mejor se ajusta a los datos es la **regresión cuadrática**. Al observar la gráfica, la curva obtenida es muy similar a la forma que siguen los datos originales, lo que indica que representa adecuadamente la relación entre las variables.

Además, este modelo obtiene un valor de \(R^2 = 0.9934\), el cual es muy alto, lo que significa que explica casi toda la variabilidad de los datos.

Al comparar con modelos de grados más altos, se observa que las gráficas y los valores de \(R^2\) no cambian de manera significativa. Por lo tanto, el modelo cuadrático es el más adecuado, ya que logra un buen ajuste sin agregar complejidad innecesaria. Esto también permite un mejor comportamiento al extrapolar, ya que los modelos de grados altos tienden a volverse inestables fuera del rango de los datos.
